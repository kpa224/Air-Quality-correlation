```python
#import necessary library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#import data from two csv files
data_daily= pd.read_csv("gazoldo_pm_1d.csv", index_col=0, parse_dates=True)
data_15min= pd.read_csv("gazoldo_weather_15_min.csv")
data_15min # view data

correlation1 = data_15min["Zona Belvedere_pm10"].corr(data_15min["Zona Belvedere_temperature"])
correlation2 = data_15min["Zona Ex Pesa_pm10"].corr(data_15min["Zona Ex Pesa_temperature"])
correlation3 = data_15min["Via Ragazzi del 99_pm10"].corr(data_15min["Via Ragazzi del 99_temperature"])
correlation4 = data_15min["Chimica Casearia_pm10"].corr(data_15min["Chimica Casearia_temperature"])
correlation =[correlation1, correlation2, correlation3, correlation4]
correlation 
# correlation for pm10 and temperature for different regions, 
#It is evident that for Zona Ex Pesa and Via ragazzi del 99 , there is inverse correlation for the two measurements

```

```python
data_daily.plot()
#It is evident from the figurethat in the month of November, there is very high air pollution index, this can be well explained by
# the correlation described before, with low temperature, the AQI is high.
plt.xlabel("Months")
plt.ylabel("AQI")
```
