# Basic dataset description

Wiseair is supporting the municipality of Gazoldo degli Ippoliti in understanding the quality of the air in their territory, in order to improve it. 


The basic dataset contains the data recorded by Wiseair sensors in Gazoldo degli Ippoliti from 2021-07-16 until 2021-12-31.


The sensors are located in the following locations (latitude North, longitude East):

1. Zona Belvedere: 45.205167061173725, 10.582332173357656
2. Zona Ex Pesa: 45.19869439610221, 10.578970961363838
3. Via Ragazzi del 99: 45.19638264035279, 10.576218027301953
4. Zona Chimica Casearia: 45.2016015, 10.5838449



The basic dataset is composed of two files, which can be downloaded in [this Google Drive folder](https://drive.google.com/drive/folders/1KYcmW6FqxRz-nx6KjT2r4guO1rtikKJd?usp=sharing).

## gazoldo_weather_15_min.csv

This file contains raw weather and pollution data, averaged every 15 minutes. The columns are (X is the name of the zone):
1. valid_at: the beginning of the 15 minutes interval, in UTC timezone, over which the data are averaged
2. X_pm1: the concentration of PM1, in ug/m3. Measured by Wiseair's sensor.
3. X_pm2p5: the concentration of PM2.5, in ug/m3. Measured by Wiseair's sensor.
4. X_pm4: the concentration of PM4, in ug/m3. Measured by Wiseair's sensor.
5. X_pm10: the concentration of PM10, in ug/m3. Measured by Wiseair's sensor.
6. X_relative_humidity: the relative rumidity, in percentage. Measured by Wiseair's sensor.
7. X_temperature: temperature in Celsius. Measured by Wiseair's sensor.
8. X_pressure: the pressure in millibar. Imported via third party API.
9. X_rain_last_1h: mm of rain fallen in the last 60 minutes. Imported via third party API.
10. X_wind_speed: wind speed in m/s. Imported via third party API.
11. X_wind_direction: wind direction in meteorological degrees. Imported via third party API.
12. X_cloud_coverage: percentage of sky covered by clouds. Imported via third party API.

## gazoldo_pm_1d.csv
This file contains daily averages of PM2.5 and PM10. These values are NOT the average of PM2.5, PM10 columns contained in the previous file. 

As a matter of fact, the accuracy of low cost PM sensors is negatively impacted by environmental factors such as high levels of relative humidity, so they need to be corrected before correctly representing the level of air pollution. 

The data in this file have been corrected using Wiseair's proprietary algorithms, so they can be considered as the real level of air pollution in the given location and day. 

The columns are (X is the name of the zone):
1. valid_at: the UTC day to which the measure refers
1. X_pm2p5: the daily average of PM2.5 concentration, in ug/m3
2. X_pm10: the daily average of PM10 concentration, in ug/m3
